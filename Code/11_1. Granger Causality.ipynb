```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

from datetime import datetime
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.api import VAR
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller, grangercausalitytests
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel('../Data/us_macro_quarterly.xlsx')
df.head()
```

```python
def parse_date(data):
    # quarter is in {01, 02, 03, 04}
    year, quarter = data.split(":")
    month = (int(quarter) -1) * 3 + 1
    s = f"{year}-{month}"
    return datetime.strptime(s, "%Y-%m")
```

```python
df['Date'] = df['Unnamed: 0'].apply(parse_date)
df.head()
```

```python
df.set_index(df['Date'], inplace = True)
df.drop(['Date', 'Unnamed: 0'], axis = 1, inplace = True)

df.index.freq = 'QS'

df.head()
```

```python
# GDP
df['GDPC96'].plot(figsize = (15,5))
```

```python
df['GDPC96 Growth'] = np.log(df['GDPC96']).diff()
df['GDPC96 Growth'].plot(figsize = (15,5))
```

```python
# There are many kinds of 'spread', but in here, using term spread 
# Term Spread = long term rate - short term rate (장기금리 - 단기 금리)
df['TSpread'] = df['GS10'] - df['TB3MS']  
df['TSpread'].plot(figsize = (15,5))
```

```python
cols = ['GDPC96 Growth', 'TSpread']

df1 = df.iloc[1:][cols].copy()
```

# Granger Causality

```python
granger_result = grangercausalitytests(df1, maxlag = 18)
# Wheter 2'nd column is affecting to 1'st -> Can TSpread help to forecast GDP Growth
```

```python
granger_result = grangercausalitytests(df1[reversed(cols)], maxlag = 18)
```
