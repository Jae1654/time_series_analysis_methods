```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools


from sklearn.metrics import mean_squared_error, r2_score
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df = pd.read_csv("../Data/sp500sub.csv",
                 index_col = 'Date',
                 parse_dates = True)
df.head()
```

```python
goog = df[df['Name'] == 'GOOG'][['Close']].copy()
goog.plot(figsize = (20, 8))
```

```python
goog['Log Close'] = np.log(goog['Close'])
goog['Log Close'].plot(figsize = (20, 8))
```

```python
# set walkforward parameters
h = 14
step = 5
Ntest = len(goog) - h - step + 1
```

```python
# Configuaration hpyerparameters to try
trend_type_list = ['add', 'mul']
damped_trend_list = [True, False]
init_method_list = ['estimated', 'heuristic', 'legacy-heuristic']
use_boxcox_list = [True, False, 0]
```

```python
def walkforward(
    trend_type,
    damped_trend,
    init_method,
    use_boxcox,
    debug = False):
    
    # store errors
    errors = []
    seen_last = False
    steps_completed = 0
    
    for end_of_train in range(Ntest, len(goog)- h + 1):
        train = goog.iloc[:end_of_train]
        test = goog.iloc[end_of_train:end_of_train + h]
        
        if test.index[-1] == df.index[-1]:
            seen_last = True
        
        steps_completed += 1
        
        # instanciate model
        hw = ExponentialSmoothing(
            train['Close'],
            initialization_method = init_method,
            trend = trend_type,
            damped_trend = damped_trend,
            use_boxcox = use_boxcox)
        
        # fit model
        res_hw = hw.fit()
        # compute error       
        fcast = res_hw.forecast(h)
        try:
            error = mean_squared_error(test['Close'], fcast)
            errors.append(error)
        except:
            continue
        
        
        
    if debug:
        print('Seen_last:', seen_last)
        print('Steps completed:', steps_completed)
            
        
    return np.mean(errors)
```

```python
best_score = float('inf')
best_options = None

tuple_of_option_lists = (trend_type_list, damped_trend_list,
                         init_method_list, use_boxcox_list)


for x in itertools.product(*tuple_of_option_lists):
    score = walkforward(*x)
    
    if score < best_score:
        print("btest score so far:", score)
        best_score = score
        best_options = x
```

```python
t, d, i , u = best_options

train = goog[:-h]


# instanciate model
hw = ExponentialSmoothing(
    train['Close'],
    initialization_method = i,
    trend = t,
    damped_trend = d,
    use_boxcox = u)
        
        # fit model
res_hw = hw.fit()
        # compute error       
fcast = res_hw.forecast(h)
```

```python
train = goog.iloc[:-h]
test = goog.iloc[-h:]

# Boolean Series to index of df rows
train_idx = goog.index <= train.index[-1]
test_idx = goog.index > train.index[-1]

goog.loc[test_idx,"Close Forecast"] = fcast.to_numpy()
```

```python
goog[["Close", "Close Forecast"]].plot()
```

# Holt-Winter vs Naive forecast

```python
naive = goog.iloc[-(h+1)]['Close']

goog.loc[test_idx,"Naive Forecast"] = naive

goog[["Close", "Naive Forecast", "Close Forecast"]].iloc[-100:].plot()


test = goog["Close"][-h:]

naive_error = mean_squared_error(test, [naive]*h)
hw_error = mean_squared_error(test, fcast)


print('Naive Forecast Error:', naive_error)
print('Holt-Winter Forecast Error:', hw_error)
```
