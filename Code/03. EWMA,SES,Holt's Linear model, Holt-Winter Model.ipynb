```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
df = pd.read_csv("../Data/airline_passengers.csv", index_col = 'Month',parse_dates = True)
df.head()
```

```python
# Check missing values
df.isna().sum()
```

```python
# Basic Plot
df.plot(figsize= (10, 5))
```

# 1. EWMA

```python
alpha = 0.2
# set adjust = False to make it EWMA
df['EWMA'] = df['Passengers'].ewm(alpha = alpha, adjust = False).mean()
```

```python
df.plot()
```

# 2. SES

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
# This gives warnning since df doens't have frequency
ses = SimpleExpSmoothing(df['Passengers'])
```

```python
# No frequency is given, hence SES gives a warnning.
df.index
```

```python
# Set frequency
df.index.freq = 'MS'
```

```python
# This gives no warnning
ses = SimpleExpSmoothing(df['Passengers'])
```

```python
alpha = 0.2
res = ses.fit(smoothing_level = alpha, optimized = False)
```

```python
df['SES'] = res.predict(start = df.index[0], end = df.index[-1])
```

```python
# check two timeseries are the same
np.allclose(df['SES'], res.fittedvalues)
```

```python
df.plot(figsize = (10,5 ))
```

```python
# The SES method seems shifted by 1 compared to EWMA
df.head()
```

## 2.1 SES Application

```python
N_test = 12
train = df.iloc[:-N_test]
test = df.iloc[-N_test:]


# Not explicitly passing alpha so that ses finds the best
ses = SimpleExpSmoothing(
    train['Passengers'])

# fit model
res = ses.fit()

# boolean series to index df rows
train_idx = df.index <= train.index[-1]
test_idx = df.index > train.index[-1]
```

```python
df.loc[train_idx, 'SESfitted'] = res.fittedvalues
df.loc[test_idx, 'SESfitted'] = res.forecast(N_test)

# Forecast is the straight horizontal line
# The fitted values seems to be lagged by 1, but it is because the model consider recent sample for prediction
df[['Passengers', 'SESfitted']].plot()
```

```python
# From the above plot, we can see that the SES model reacts to the
#  present value very quickly meaning that alpha is close to 1
res.params
```

# Holt-Winter 

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Additive model
hw = ExponentialSmoothing(train['Passengers'],
                         trend = 'add',
                         seasonal = 'add',
                         seasonal_periods = 12)

res_hw = hw.fit()

df.loc[train_idx, 'Holt-Winters'] = res_hw.fittedvalues
df.loc[test_idx, 'Holt-Winters'] = res_hw.forecast(N_test)

# Model fits really well
df[['Passengers', 'Holt-Winters']].plot()
```

```python
def rmse(y, t):
    return np.sqrt(np.mean(y-t)**2)

def mae(y, t):
    return np.mean(np.abs(y-t))
```

```python

print("Train RMSE:", rmse(train['Passengers'], res_hw.fittedvalues))
print("Test RMSE:", rmse(test['Passengers'], res_hw.forecast(N_test)))


print("Train MAE:", mae(train['Passengers'], res_hw.fittedvalues))
print("Test MAE:", mae(test['Passengers'], res_hw.forecast(N_test)))
```

```python
# Trend additive, Seaosnal Multiplicative model
hw = ExponentialSmoothing(train['Passengers'],
                         trend = 'add',
                         seasonal = 'mul',
                         seasonal_periods = 12)

res_hw = hw.fit()

df.loc[train_idx, 'Holt-Winters'] = res_hw.fittedvalues
df.loc[test_idx, 'Holt-Winters'] = res_hw.forecast(N_test)

# Model fits really well
df[['Passengers', 'Holt-Winters']].plot()
```

```python
print("Train RMSE:", rmse(train['Passengers'], res_hw.fittedvalues))
print("Test RMSE:", rmse(test['Passengers'], res_hw.forecast(N_test)))


print("Train MAE:", mae(train['Passengers'], res_hw.fittedvalues))
print("Test MAE:", mae(test['Passengers'], res_hw.forecast(N_test)))
```

```python
# Multiplicative model because trends seems growing
hw = ExponentialSmoothing(train['Passengers'],
                         trend = 'mul',
                         seasonal = 'mul',
                         seasonal_periods = 12)

res_hw = hw.fit()

df.loc[train_idx, 'Holt-Winters'] = res_hw.fittedvalues
df.loc[test_idx, 'Holt-Winters'] = res_hw.forecast(N_test)

# Model fits really well
df[['Passengers', 'Holt-Winters']].plot()
```
